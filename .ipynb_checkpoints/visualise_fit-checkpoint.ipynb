```python
params = ['omega_cdm', 'omega_b', 'h', 'ln10^{10}A_s', 'n_s',
                              'b_1', 'b_2', 'bG2', 'bGamma3', 'c0', 'c2',  
                             'ch', 'Pshot', 'a0']
post1d = {}
max_post = {}
for i,p in enumerate(params):
    kde = gaussian_kde(r['Pkcosmo']['samples'][:, i])
    x = np.linspace(np.min(r['Pkcosmo']['samples'][:, i]), np.max(r['Pkcosmo']['samples'][:, i]), 200)
    post1d[p] = [x,kde(x)]
    max_post[p] = x[np.argmax(kde(x))]
```
