```python
chain_b1 = np.concatenate([np.load(fn_b1[i]) for i in range(2)])
chain_biases = np.concatenate([np.load(fn_biases[i]) for i in range(2)])
chain_ocdm = np.concatenate([np.load(fn_ocdm[i]) for i in range(2)])
chain_cosmo = np.concatenate([np.load(fn_cosmo[i]) for i in range(2)])
chain_b1_pk = np.concatenate([np.load(fn_b1_pk[i]) for i in range(2)])
chain_biases_pk = np.concatenate([np.load(fn_biases_pk[i]) for i in range(2)])
chain_ocdm_pk = np.concatenate([np.load(fn_ocdm_pk[i]) for i in range(2)])
```
