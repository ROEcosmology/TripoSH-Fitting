```python
from classy import Class
import numpy as np
from matplotlib import pyplot as plt
from fit_test import rebuild_pk_class_pt
```

```python
h_fid = .6736
omega_b_fid = .02237
omega_cdm_fid = .12
n_s_fid = .9649
A_s_fid = 2.083e-9

z = .8

common_settings = {
            'output': 'mPk',
            'h': h_fid,
            'omega_b': omega_b_fid,
            'omega_cdm': omega_cdm_fid,
            'n_s': n_s_fid,
            'A_s': A_s_fid,
            'tau_reio': 0.0544,
            'N_ncdm': 1.,
            'm_ncdm': 0.06,
            'N_ur': 2.0328,
            'z_max_pk': 3.,
            'P_k_max_h/Mpc': 50.,
            'z_pk':z} 

M = Class()
M.set(common_settings)
M.set({'output':'mPk',
        'non linear':'PT',
        'IR resummation':'Yes',
        'Bias tracers':'Yes',
        'cb':'Yes',
        'RSD':'Yes',
       })
M.compute()
```

```python
k=np.arange(.0025, .3, .0025)

kh = k*h_fid

fz = M.scale_independent_growth_factor_f(z)
M.initialize_output(kh, z, len(kh))
M_mult = M.get_pk_mult(kh, z, len(kh))
```

```python
bias_params = {'b1': 1.9943274e+00,
 'b2': -2.8493848e-01,
 'bG2': -2.9738286e-01,
 'bGamma3': 6.8452853e-01,
 'c0': 1.1104226e+01,
 'c2pp': 3.0399105e+01,
 'c4pp': 3.0399105e+01,
 'ch': 3.9950281e+02,
 'Pshot': 5.6773871e-01,
 'a0': -4.3189287e+00,
                'f': fz}
```

```python
Pshot_norm = (1 + bias_params['Pshot'] + bias_params['a0']*k**2) / 1e-3
```

```python
pk = rebuild_pk_class_pt(M_mult, bias_params, fz, h_fid, k, ells=[0,2,4])
#pk[0] += Pshot_norm
```

```python
pk_model_classpt = {}
pk_model_classpt[0] = M.pk_gg_l0(bias_params['b1'],bias_params['b2'],bias_params['bG2'],bias_params['bGamma3'],bias_params['c0'],
                              0,bias_params['ch'])
#pk_model_classpt[0] += Pshot_norm
pk_model_classpt[2] = M.pk_gg_l2(bias_params['b1'],bias_params['b2'],bias_params['bG2'],bias_params['bGamma3'],bias_params['c2pp'],
                              bias_params['ch'])
pk_model_classpt[4] = M.pk_gg_l4(bias_params['b1'],bias_params['b2'],bias_params['bG2'],bias_params['bGamma3'],bias_params['c4pp'],
                              bias_params['ch'])
```

```python
for ell in [0,2,4]:
    plt.plot(k, k*pk[ell], label=ell)
    plt.plot(k, k*pk_model_classpt[ell], label=f'cpt {ell}')
plt.legend()
```

```python
for ell in [0,2,4]:
    plt.plot(k, pk[ell]/pk_model_classpt[ell], label=ell)
    plt.show()
```
