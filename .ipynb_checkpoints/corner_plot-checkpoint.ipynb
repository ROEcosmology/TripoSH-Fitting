```python
import os
from classy import Class
import numpy as np
from matplotlib import pyplot as plt
from fit import *
```

```python
s = {}
s['pk'] = pc.Sampler(1000,2,loglike_poco,logprior_poco)
s['bk'] = pc.Sampler(1000,2,loglike_poco,logprior_poco)
s['pbk'] = pc.Sampler(1000,2,loglike_poco,logprior_poco)
```

```python
r = {}
for est in ['pk','bk','pbk']:
    if est=='pbk': fil = ''
    else: fil = '_'+est
    s[est].load_state(path=f'/home/rneveux/fit_test/test{fil}.state')
    r[est] = s[est].results
```

```python
fig = pc.plotting.corner(r['pk'], labels=[r'$b_1$', r'$b_2$', r'$b_{G_2}$', r'$b_{\Gamma 3}$'], smooth=True, smooth1d=True, plot_density=False, 
                  plot_datapoints=False, levels=[.67, .95], color='r')
pc.plotting.corner(r['bk'], labels=[r'$b_1$', r'$b_2$', r'$b_{G_2}$', r'$b_{\Gamma 3}$'], smooth=True, smooth1d=True, plot_density=False, 
                  plot_datapoints=False, levels=[.67, .95], color='b',fig=fig)
pc.plotting.corner(r['pbk'], labels=[r'$b_1$', r'$b_2$', r'$b_{G_2}$', r'$b_{\Gamma 3}$'], smooth=True, smooth1d=True, plot_density=False, 
                  plot_datapoints=False, levels=[.67, .95], color='k',fig=fig)
```

```python
for est in ['pk','bk','pbk']:
    print(est)
    print('Mean values = ', np.mean(r[est]['samples'], axis=0))
    print('Standard deviation values = ', np.std(r[est]['samples'], axis=0))
```
